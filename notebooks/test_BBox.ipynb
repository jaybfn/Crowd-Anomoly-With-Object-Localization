```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import cv2

im = Image.open('../data/ucf_action/Kicking/001/6729-10_70335.jpg')

# Create figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(im)

# Create a Rectangle patch
rect = patches.Rectangle((540, 212), 65, 80, linewidth=1, edgecolor='r', facecolor='none')

# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
    #plt.show() 
```
